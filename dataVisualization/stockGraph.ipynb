```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [2, 3, 5, 7, 11]
})

# Histogram
sns.histplot(data['A'], bins=5, kde=True)
plt.title('Histogram of A')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Scatter plot
sns.scatterplot(x='A', y='B', data=data)
plt.title('Scatter Plot of A vs. B')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```
